```python
import numpy as np
import matplotlib.pyplot as plt

def cepstrum_analysis(signal, fs):
    # 푸리에 변환
    spectrum = np.fft.fft(signal)
    magnitude = np.abs(spectrum)
    
    # 로그 스펙트럼
    log_spectrum = np.log(magnitude + 1e-10)  # 작은 값 추가하여 로그 계산 시 0 방지
    
    # 역푸리에 변환
    cepstrum = np.fft.ifft(log_spectrum).real
    
    return cepstrum

# 예시 신호 생성 (사인파)
fs = 8000  # 샘플링 주파수
t = np.linspace(0, 1, fs)
frequency = 440  # 주파수 (A4)
signal = 0.5 * np.sin(2 * np.pi * frequency * t)

# 켑스트럼 분석 수행
cepstrum_result = cepstrum_analysis(signal, fs)

# 결과 시각화
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.title("Original Signal")
plt.plot(t, signal)
plt.subplot(2, 1, 2)
plt.title("Cepstrum")
plt.plot(cepstrum_result)
plt.tight_layout()
plt.show()
```
